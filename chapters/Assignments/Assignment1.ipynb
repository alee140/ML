```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tseries.offsets import MonthEnd

# your code below
#URL of the website & sheetname '49_Industry_Portfolios'
url="https://github.com/amoreira2/Fin418/blob/main/assets/data/Assignment1.xlsx?raw=true"
sheet_name = "49_Industry_Portfolios"

#start reading data from row 7 but using 6 because count starts from 0
skip_rows = 6


#import data as dataframe df_ind AND replace -99.99 and -999 with NaN

df_ind = pd.read_excel(url, sheet_name=sheet_name,skiprows=skip_rows, usecols='A:AX', na_values=['N/A', -999, -99.99])


##change the first column which is 'Unnamed: 0' to date
df_ind = df_ind.rename(columns={'Unnamed: 0':'date'})


##change the date column to datetime objects
df_ind['date'] = pd.to_datetime(df_ind['date'], format='%Y%m', errors='coerce')


# Set date as index
df_ind.set_index('date', inplace=True)


print(df_ind.head())
df_ind.info()










```
